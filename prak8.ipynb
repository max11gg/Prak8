Цель: Ознакомится с моделями классификации и регрессии.

На данном этапе здесь будут подключаться все необходимые библиотеки для работы с датасетом и варинт кодирования категориальной переменной, и для разбивки данных.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import KFold
```

С помощью функции read_csv происходит подключение и чтение нового файла с работой "vgsales_51.csv", который не содержит пустых строк, сделанной в работе №1.

```python
df = pd.read_csv("vgsales_51.csv")
df
```

Затем необходимо все категориальные значения из набора данных преобразовать в бинарные, для этого будем использовать кодирование из прошлой практической работы - OrdinalEncoder, а столбец Genre преобразовать через OneHotEncoder, потому что там меньше всего уникальных значений. Через fit_transform преобразовываем старые данные в новые.

```python
OE = OrdinalEncoder()
df[['Name', 'Platform', 'Year', 'Publisher']] = OE.fit_transform(df[['Name', 'Platform', 'Year', 'Publisher']])
OHE = OneHotEncoder()
df['Genre'] = OHE.fit_transform(df[['Genre']]).toarray()
df
```

k-блочная кросс-валидация – процедура, используемая для оценки навыков Моделей Машинного обучения на новых данных. Она обычно используется для сравнения и выбора модели конкретной задачи прогнозного моделирования, потому что его легко понять и реализовать.
В ней используются параметры: n_splits=5 (Количество сгибов, затем каждое сгибание используется один раз в качестве проверки), shuffle=False(Следует ли перемешивать данные перед разделением на пакеты), random_state=None(Когда shuffle верно, random_state влияет на порядок индексов, который контролирует случайность каждого сгиба. В противном случае этот параметр не будет действовать)

```python
Y = df["Genre"]
X = df.drop(columns = "Genre")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

Метод split () в Python разделяет строку на список подстрок по разделителю.

```python
kf.split(df)
```

Здесь решил выполнить одно и тоже действие двумя способами. Вначале подключаются и объявляются переменные для обучения моделей, в моем случае 3: к-ближайших соседей и метода опорных векторов и дерева решений.

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
from sklearn.svm import SVC
svc = SVC()
```

Этот фрагмент кода демонстрирует цикл for, использующий функцию train_test_split для выполнения k-кратной перекрестной валидации на наборе данных и выполнения обучения и предсказания различных варинтов моделей.

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]

    knc.fit(X_train_kfold,Y_train_kfold)
    predict_knc = knc.predict(X_test_kfold)

    dtc = dtc.fit(X_train_kfold,Y_train_kfold)
    predict_dtc = dtc.predict(X_test_kfold)

    svc.fit(X_train_kfold,Y_train_kfold)
    predict_svc = svc.predict(X_test_kfold)
```

Теперь напишу тоже самое но слегка по-другому. Здесь обучением и предсказаваем по способу К-ближайших соседей.

```python

from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()

knc.fit(X_train_kfold,Y_train_kfold)
predict_knc = knc.predict(X_test_kfold)
```

Тут подключаем и записываем в переменную споособ обучения - Дерева решений.

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()

dtc = dtc.fit(X_train_kfold,Y_train_kfold)
predict_dtc = dtc.predict(X_test_kfold)
```

По заданию необходимо сделать от 2-4 модели обучения, поэтому сделаю еще одну третью с помощью обучения метода опорных векторов.

```python
from sklearn.svm import SVC
svc = SVC()

svc.fit(X_train_kfold,Y_train_kfold)
predict_svc = svc.predict(X_test_kfold)
```

И для вывода результатов используется print с выводом как раз-таки этих обученных моделей.

```python
print("\n")
print("Набор предсказаний дерева решений: ",predict_dtc)
print("Набор предсказаний К-ближайших соседей: ",predict_knc)
print("Набор предсказаний метода опорных векторов: ",predict_svc)
```

Регрессия – группа Моделей, используемых для прогнозирования непрерывных значений, таких как цены на недвижимость с учетом их характеристик. Для регресии будут использоваться тот же k-блочная кросс-валидация, как и было продемонстрировано выше для сравнения и выбора модели конкретной задачи прогнозного моделирования

```python
Y = df["Global_Sales"]
X = df.drop(columns = "Global_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

Метод split () в Python разделяет строку на список подстрок по разделителю.

```python
kf.split(df)
```

Здесь точно также будет использоваться одно и тоже, но слегка по-разному. Подключаем библиотеки и записываем их в переменные для регрессии опорных векторов, регрессии дерева решений и линейной регрессии.

```python
from sklearn.linear_model import LinearRegression
lrc = LinearRegression()
from sklearn.svm import SVR
svr = SVR()
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

Этот фрагмент кода демонстрирует цикл for, использующий функцию train_test_split для выполнения k-кратной перекрестной валидации на наборе данных и выполнения обучения и предсказания различных варинтов моделей регрессии.

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]

    lrc.fit(X_train_kfold,Y_train_kfold)
    predict_lrc = lrc.predict(X_test_kfold)

    svr.fit(X_train_kfold,Y_train_kfold)
    predict_svr = svr.predict(X_test_kfold)

    dtr = dtr.fit(X_train_kfold,Y_train_kfold)
    predict_dtr = dtr.predict(X_test_kfold)
    
```

Теперь еще вариант записать все вне кода с циклом. Для этого записываем и указываем обучение линейной регрессии.

```python
from sklearn.linear_model import LinearRegression
lrc = LinearRegression()

lrc.fit(X_train_kfold,Y_train_kfold)
predict_lrc = lrc.predict(X_test_kfold)
```

Здесь происходит обучение и предсказание Регрессии опорных векторов.

```python
from sklearn.svm import SVR
svr = SVR()

svr.fit(X_train_kfold,Y_train_kfold)
predict_svr = svr.predict(X_test_kfold)
```

И последней выбранной моделью по заданию будет регрессия дерева решений.

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()

dtr = dtr.fit(X_train_kfold,Y_train_kfold)
predict_dtr = dtr.predict(X_test_kfold)
```

И точно также для проверки выведенных регресии моделей произведем вывод этих значений.

```python
print("\n")
print("Набор предсказаний регрессии дерева решений: ",predict_dtr)
print("Набор предсказаний регрессии опорных векторов: ",predict_svr)
print("Набор предсказаний линейной регрессии: ",predict_lrc)
```
